```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("train.csv")

plt.figure(figsize=(12, 8))
plt.imshow(data.corr(numeric_only=True), cmap='Blues', interpolation='none', aspect='auto')
plt.colorbar()
plt.xticks(range(len(data.corr(numeric_only=True))), data.corr(numeric_only=True).columns, rotation='vertical')
plt.yticks(range(len(data.corr(numeric_only=True))), data.corr(numeric_only=True).columns)
plt.title("Тепловая карточенька корреляций между признаками")
plt.show()
```
